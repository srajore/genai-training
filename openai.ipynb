```python
from openai import OpenAI
```

```python
client = OpenAI(api_key="")
```

```python
prompt= "What is GenAI ,explain in simple term?"
```

```python
completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ]
)
```

```python
print(completion.choices[0].message.content)
```
